```python
#Importamos las librerias que vamos a utilizar
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from ydata_profiling import ProfileReport
```

```python
# Cargamos el dataset
df_main = pd.read_csv('C:/Users/tomif/Desktop/Proyecto_1/Datasets utilizados/Main.csv')
```

```python
# Visualización de Outliers

# Elegimos las columnas numericas de interés en df_peliculas
columnas_numericas_peliculas = df_main.select_dtypes(include=['float64', 'int64', 'int32']).columns


# Creamos un boxplots para todas las columnas numéricas en df_peliculas usando escala logarítmica
plt.figure(figsize=(15, 10))
df_main[columnas_numericas_peliculas].apply(lambda x: np.log1p(x)).boxplot()
plt.title('Boxplots de todas las columnas numéricas en df_peliculas (escala logarítmica)')
plt.xticks(rotation=90)
plt.show()
```
