```python
#Importamos las librerias que vamos a utilizar
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from ydata_profiling import ProfileReport
```

```python
# Cargamos el dataset
df_main = pd.read_csv('C:/Users/tomif/Desktop/Proyecto_1/Datasets utilizados/Main.csv')
```

```python
# Visuañizacion de los valores nulos en el dataset:

# Verificamos los valores nulos
null_values = df_main.isnull().sum()

# Creamos un dataframe para facilitar el gráfico
null_values_df = pd.DataFrame(null_values, columns=['nulos'])
null_values_df = null_values_df.reset_index()
null_values_df.columns = ['Variable', 'Nulos']

# Creamos el gráfico
plt.figure(figsize=(12, 8))
sns.barplot(x='Variable', y='Nulos', data=null_values_df, palette='viridis')
plt.xticks(rotation=90)
plt.title('Valores nulos por variable')
plt.ylabel('Cantidad de valores nulos')
plt.xlabel('Variables')
plt.show()
```

```python
#Visualizamos la distribución de las variables numéricas

df_main.hist(figsize=(14, 10), bins=20, color='blue', edgecolor='black')
plt.suptitle('Distribución de variables numéricas')
plt.show()
```

```python
# Visualización de Outliers

# Elegimos las columnas numericas de interés en df_peliculas
columnas_numericas_peliculas = df_main.select_dtypes(include=['float64', 'int64', 'int32']).columns


# Creamos un boxplots para todas las columnas numéricas en df_peliculas usando escala logarítmica
plt.figure(figsize=(15, 10))
df_main[columnas_numericas_peliculas].apply(lambda x: np.log1p(x)).boxplot()
plt.title('Boxplots de todas las columnas numéricas en df_peliculas (escala logarítmica)')
plt.xticks(rotation=90)
plt.show()
```
